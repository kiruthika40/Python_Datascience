```python
import requests
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population').text
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
print('\n')
My_table = soup.find('table',{'class':'wikitable'})
# My_table
rows = My_table.select('tbody > tr')
header = [th.text.rstrip() for th in rows[0].find_all('th')]  
print("#######Header#######")
print(header)
allRows = My_table.tbody.findAll('tr')
allRows = allRows[1:10]
countries = []
population = []
for row in allRows[1:]:
        countries.append(row.find('a').text)
        population.append(int(row.findAll('td')[1].text.replace(',','')))
            

print('\n')
print("********Countries list*********")
for c in countries:
    print(c)
print('\n')
print("********Population List*********")
for pop in population:
    print(pop)
print('\n')
from matplotlib import pyplot as plt 

fig = plt.figure(figsize =(10, 7)) 
plt.legend(population, countries,
          title="Countries and Population: A pie",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))


plt.pie(population, labels = countries,autopct='%1.1f%%',pctdistance=1.25, labeldistance=.6) 
```
